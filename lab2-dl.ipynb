```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=x*2
x.dtype
```

```python
print(x)
print(y)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)
```

```python
w1 = np.random.rand()   
b1 = np.random.rand()   

w2 = np.random.rand()   
b2 = np.random.rand()   

eta = 0.01              
epochs = 1000

loss_list = []
```

```python
for epoch in range(1000):
    total_loss = 0
    for i in range(len(x)):
        z1 = w1 * x[i] + b1
        a1 = sigmoid(z1)

        z2 = w2 * a1 + b2
        a2 = sigmoid(z2)

        error = y[i] - a2
        loss = 0.5 * error**2
        total_loss += loss

        delta2 = error * sigmoid_derivative(z2)
        delta1 = delta2 * w2 * sigmoid_derivative(z1)


        w2 = w2 + eta * delta2 * a1
        b2 = b2 + eta * delta2

        w1 = w1 + eta * delta1 * x[i]
        b1 = b1 + eta * delta1

    loss_list.append(total_loss)
```

```python
x_test = 6

z1 = w1 * x_test + b1
a1 = sigmoid(z1)

z2 = w2 * a1 + b2
y_pred = sigmoid(z2)

print("Test input:", x_test)
print("Predicted output:", y_pred)
```

```python
plt.plot(loss_list)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss vs Epochs")
plt.show()
```
